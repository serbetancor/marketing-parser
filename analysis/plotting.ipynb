```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('../data/finalZara.json', 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data['categories'])
```

```python
statistics = df.describe()
total_products = df['quantity'].sum()
average_price = df['avg_price'].mean()
```

```python
# Gráfico de barras de la cantidad de productos por categoría
plt.figure(figsize=(12, 6))
plt.bar(df['name'], df['quantity'])
plt.xlabel('Categoría')
plt.ylabel('Cantidad de Productos')
plt.title('Cantidad de Productos por Categoría')
plt.xticks(rotation=45)
plt.show()

# Gráfico de pastel del promedio de precios por categoría
plt.figure(figsize=(8, 8))
plt.pie(df['avg_price'], labels=df['name'], autopct='%1.1f%%')
plt.title('Promedio de Precios por Categoría')
plt.show()
```
