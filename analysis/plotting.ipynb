```python
import json
import pandas as pd
import matplotlib.pyplot as plt
```

```python
with open('../data/finalZara.json', 'r') as f:
    data = json.load(f)

df = pd.DataFrame(data['categories'])
```

```python
statistics = df.describe()
total_products = df['quantity'].sum()
average_price = df['avg_price'].mean()
```

```python
# Grfico de barras de la cantidad de productos por categora
plt.figure(figsize=(12, 6))
plt.bar(df['name'], df['quantity'])
plt.xlabel('Categora')
plt.ylabel('Cantidad de Productos')
plt.title('Cantidad de Productos por Categora')
plt.xticks(rotation=45)
plt.show()

# Grfico de pastel del promedio de precios por categora
plt.figure(figsize=(8, 8))
plt.pie(df['avg_price'], labels=df['name'], autopct='%1.1f%%')
plt.title('Promedio de Precios por Categora')
plt.show()
```
